```python
# @title Fun√ß√µes
def agenda():
  try:
    selecao = int(input('''
    üë• Agenda de Contatos:
    ‚ùñ  Menu:

    1 - Adicionar Contato
    2 - Buscar Contato
    3 - Listar Contatos
    4 - Editar Contato
    5 - Remover Contato
    6 - Sair
    ‚ñ∂ Escolha uma op√ß√£o: '''))

    if 1 <= selecao <= 6:
      return selecao
    else:
      print('Op√ß√£o inv√°lida. Por favor, escolha um n√∫mero entre 1 e 6.')
  except ValueError:
    print('Op√ß√£o inv√°lida. Por favor, escolha um n√∫mero entre 1 e 6.')
  return None

def validar_telefone(telefone):
    '''Valida se o telefone est√° no formato (xx)xxxx-xxxx ou xxxx-xxxx'''
    telefone = telefone.strip()

    # Verifica o formato xxxx-xxxx
    if len(telefone) == 9:  # xxxx-xxxx
      if telefone[4] == '-':  # verifica se tem h√≠fen na posi√ß√£o correta
          parte1 = telefone[:4]
          parte2 = telefone[5:]
          if parte1.isdigit() and parte2.isdigit():
            return True

    # Verifica o formato (xx)xxxx-xxxx
    elif len(telefone) == 13:  # (xx)xxxx-xxxx
      if telefone[0] == '(' and telefone[3] == ')' and telefone[8] == '-':
          ddd = telefone[1:3]
          parte1 = telefone[4:8]
          parte2 = telefone[9:]
          if ddd.isdigit() and parte1.isdigit() and parte2.isdigit():
            return True
    return False

def validar_email(email):
  '''Valida se o e-mail tem o formato b√°sico de email@dominio.com'''
  if '@' in email and '.' in email.split('@')[1]:
    return True
  return False

############################## 1 - Adicionar Contato
def adicionar_contato(nome, telefone, email,agenda):
  '''Adiciona um novo contato √† lista de contatos'''
  if not validar_telefone(telefone):
      print('Telefone inv√°lido!')
      return
  if not validar_email(email):
      print('E-mail inv√°lido!')
      return

  if not agenda:
      agenda = {'Nome': [], 'Telefone': [], 'E-mail': []}

  agenda.setdefault('Nome', []).append(nome)
  agenda.setdefault('Telefone', []).append(telefone)
  agenda.setdefault('E-mail', []).append(email)

  print(f'Contato {nome} adicionado com sucesso!')
  return agenda

############################## 2 - Bsucar Contato | -
def buscar_contato(nome_,agenda):
  lista_contato = {}
  indice_nome = []
  indice= 0

  while indice < len(agenda['Nome']):
    if nome_ in agenda['Nome'][indice]:
      indice_nome.append(indice)
    indice +=1

  if not indice_nome:
    print(f'Nome n√£o encontrado. Digite corretamente.')

  for chave in agenda:
    lista_contato[chave] = []
    for indice_filtro in indice_nome:
      if indice_filtro < len(agenda[chave]):
        lista_contato[chave].append(agenda[chave][indice_filtro])

  return dict(lista_contato)

############################## 4 - Editar Contato
def editar_contatos(agenda, menu):
  if not agenda['Nome']:
    print('Nenhum contato existente na agenda')
    return menu

  nome_original = input('Digite o nome do contato que deseja editar (ou 0 para sair): ')

  if nome_original == '0':
    print('Edi√ß√£o cancelada.')
    return menu

  if nome_original not in agenda['Nome']:
    print('Contato n√£o encontrado, digite o nome corretamente.')
    return menu

  indice_edicao = agenda['Nome'].index(nome_original)

  while True:
    print('\nO que deseja editar?')
    print('1 - Nome')
    print('2 - Telefone')
    print('3 - E-mail')
    print('4 - Editar tudo')
    print('0 - Sair da edi√ß√£o')
    opcao = int(input('Escolha uma op√ß√£o: '))

    if opcao == 0:
      print('Edicao finalizada')
      return menu

    if opcao == 1 or opcao == 4:
      nome_novo = input('\nDigite aqui o novo nome: ')
      agenda['Nome'][indice_edicao] = nome_novo
      print('‚úî Nome atualizado com sucesso.')
      return menu

    if opcao == 2 or opcao == 4:
      novo_telefone = input('\nDigite o novo telefone ("(XX)XXXX-XXXX" ou "XXXX-XXXX"): ')
      if validar_telefone(novo_telefone):
        agenda['Telefone'][indice_edicao] = novo_telefone
        print('‚úî Telefone atualizado com sucesso!')
        return menu
      else:
        print('Telefone inv√°lido! Edi√ß√£o cancelada.')
        return menu

    if opcao == 3 or opcao == 4:
      novo_email = input('\nDigite o novo e-mail: ')
      if validar_email(novo_email):
        agenda['E-mail'][indice_edicao] = novo_email
        print('‚úî E-mail atualizado com sucesso!')
        return menu
      else:
        print('E-mail inv√°lido! Edi√ß√£o cancelada.')
        return menu

  print('\n Contato atualizado com sucesso')
  return agenda

############################## 5 - Remover Contato
def remover_contatos(agenda, menu):
  if not agenda['Nome']:
    print('Nenhum contato existente na agenda')
    return menu

  nome_original = input('Digite o nome do contato que deseja remover (ou 0 para sair): ')

  if nome_original == '0':
    print('Remo√ß√£o cancelada.')
    return menu

  if nome_original not in agenda['Nome']:
    print('Contato n√£o encontrado, digite o nome corretamente.')
    return menu

  indice_remocao = agenda['Nome'].index(nome_original)

  while True:
    print('\nO que deseja remover?')
    print('1 - Remover contato')
    print('0 - Voltar ao menu principal')
    opcao = int(input('Escolha uma op√ß√£o: '))

    if opcao == 0:
      print('Edi√ß√£o finalizada.')
      break

    if opcao == 1:
      nome_removido = input('\nRemover o nome: ')
      del agenda['Nome'][indice_remocao]
      del agenda['Telefone'][indice_remocao]
      del agenda['E-mail'][indice_remocao]
      print('‚úî Telefone removido com sucesso!')
      return menu
    else:
      print('Telefone inv√°lido! Edi√ß√£o cancelada.')
      return menu

  print('\n Contato atualizado com sucesso')
  return agenda
```
